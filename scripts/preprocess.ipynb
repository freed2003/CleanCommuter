```python
import numpy as np
import pandas as pd
```

```python
cols = ['Make','Model','CO2 Emissions(g/km)']
emissions = pd.read_csv("emissions.csv", usecols=cols)
emissions
```

```python
filter_emissions = pd.DataFrame(columns=['Make','Model','CO2 Emissions(g/km)'])
```

```python
for model in emissions.Model.unique():
    cars = emissions.loc[emissions['Model'] == model]
    avg_emit = cars['CO2 Emissions(g/km)'].mean()
    
    new_df = {'Make': cars['Make'].iloc[0], "Model": cars['Model'].iloc[0], "CO2 Emissions(g/km)": avg_emit}

    filter_emissions = filter_emissions.append(new_df, ignore_index = True)
filter_emissions
```

```python
market_cols = ['Make','Model','Year','MSRP']
market_vals = pd.read_csv("market.csv", usecols=market_cols)
market_vals
```

```python
filter_model = pd.DataFrame(columns=['Make','Model','MSRP'])
```

```python
for model in filter_year.Model.unique():
    cars = filter_year.loc[filter_year['Model'] == model]
    avg_msrp = cars['MSRP'].mean()
    
    new_df = {'Make': cars['Make'].iloc[0], "Model": cars['Model'].iloc[0], "MSRP": avg_msrp}

    filter_model = filter_model.append(new_df, ignore_index = True)
filter_model
```

```python
market_models = filter_model.Model.unique()
emit_models = filter_emissions.Model.unique()
intersect_model = []
for mm in market_models:
    for em in emit_models:
        if mm == em:
            intersect_model.append(mm)
print(len(intersect_model))
```

```python
result = pd.DataFrame(columns=['Make','Model','MSRP','CO2 Emissions(g/km)'])
```

```python
for model in intersect_model:
    e_cars = filter_emissions.loc[filter_emissions['Model'] == model]

    m_cars = filter_year.loc[filter_year['Model'] == model]
    
    new_entry = {'Make': m_cars['Make'].iloc[0], "Model": m_cars['Model'].iloc[0], "MSRP": m_cars['MSRP'].iloc[0], 'CO2 Emissions(g/km)': e_cars['CO2 Emissions(g/km)'].iloc[0]}

    result = result.append(new_entry, ignore_index = True)
result
```
